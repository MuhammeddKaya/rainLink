```python
# veri setindeki colonlar için frequency plotları çizelim

for column in df.columns:
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df, x=column)
    plt.title(f'Frequency Plot for {column}')
    plt.xticks(rotation=45)
    plt.show()
```
